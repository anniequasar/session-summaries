```
from fastai.vision.all import *
```

```
os.listdir()
```

```
path = Path('train/')
```

```
path.ls()
```

```
tfms = aug_transforms(flip_vert=False, max_lighting=0.4, max_zoom=0.9, max_rotate=20, size=224)
```

```
df.head(2)
```

```
path
```

```
dblock = DataBlock(blocks=(ImageBlock, RegressionBlock),
                get_x=ColReader(0,pref=str(path/'image_data') + "/"),
                get_y=ColReader(1),
                splitter=RandomSplitter(seed=42),
                item_tfms=Resize(800),
                batch_tfms=tfms
                )
```

```
dls = dblock.dataloaders(df, bs=64, path=path)
```

```
dls.show_batch(max_n=4)
```

```
dblock.summary(df)
```

```
dls.train.show_batch(unique=True)
```

```
learn = cnn_learner(dls, resnet34, metrics=[rmse,mae])
```

```
learn.loss_func
```

```
learn.lr_find()
```

```
learn.fine_tune(4, base_lr=1e-2)
```

```
learn.show_results()
```

```
learn.export()
```
