# Plotting

```python
x = [1,2,3,4,5]
y = [1,2,3,4,5]
```

```python
#plt.figure(figsize = (10,10))
plt.plot(x,y)
```

```python
plt.plot(x,y, 'o')
```

```python
plt.plot(x,y,'o')
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("Basic chart")
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(1,2,1)
ax2 = fig.add_subplot(1,2,2)
ax1.plot(x,y)
ax2.plot(x,y,'o')
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(2,1,1)
ax2 = fig.add_subplot(2,1,2)
ax1.plot(x,y)
ax2.plot(x,y,'o')
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(2,1,1)
ax2 = fig.add_subplot(2,1,2)
ax1.plot(x,y)
ax2.plot(x,y,'o', c = 'red')
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(2,1,1)
ax2 = fig.add_subplot(2,1,2)
ax1.plot(x,y)
ax1.set_xlabel("X axis")
ax1.set_ylabel("Y axis")
ax1.set_title("Plot 1")
ax2.plot(x,y,'o', c = 'red')
ax2.set_xlabel("X axis")
ax2.set_ylabel("Y axis")
ax2.set_title("Plot 2")
fig.tight_layout()
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(3,1,1)
ax2 = fig.add_subplot(3,1,2)
ax3 = fig.add_subplot(3,1,3)
ax1.plot(x,y)
ax1.set_xlabel("X axis")
ax1.set_ylabel("Y axis")
ax1.set_title("Plot 1")
ax2.plot(x,y,'o', c = 'red')
ax2.set_xlabel("X axis")
ax2.set_ylabel("Y axis")
ax2.set_title("Plot 2")
ax3.plot(x,y,'o', c = 'yellow')
ax3.set_xlabel("X axis")
ax3.set_ylabel("Y axis")
ax3.set_title("Plot 2")
fig.tight_layout()
```

# Bar plot 

```python
fig = plt.figure()
x = [1,2,3,4,5]
y = [90,88,20,77,45]
plt.bar(x,y)
plt.title("Bar plots")
plt.xticks(x,["A","B","C","D","E"]);
plt.yticks([20,40,60,80], ["R1","R2","R3","R4"])
fig.tight_layout()
```

```python
x = [1,2,3,4,5]
y = [90,88,20,77,45]
plt.barh(x,y)
```

# Seaborn plots
- sns.boxplot(x = .., y = .. , data = ..)
- sns.violinplot(....)
- sns.stripplot(...)
- sns.swarmplot(...)

```python
sns.boxplot(x = "Origin", y = "Horsepower", data = data)
```

```python
sns.stripplot(x = "Origin", y = "Horsepower", data = data)
```

```python
sns.swarmplot(x = "Origin", y = "Horsepower", data = data)
```

```python
sns.violinplot(x = "Origin", y = "Horsepower", data = data)
```
