Seaborn uses matplotlib as the base.
<br>
Using seaborn, you can map univariate (one variable) data by using functions such as distribution plots 

```python
sns.distplot(cdata.MPG);
```

MPG = miles per gallon

```python
sns.distplot(cdata["Year"]);
```

And what if you wanted to plot one variable against another? that would be called bivariate plotting.
<br>
You can also plot more than 2 variables on the same plot; this is called multivariate plotting. 
<br>

```python
x1 = cdata.Weight 
x2 = cdata.Horsepower
```

Using matplotlib to plot bivariate data, we get something like this:

```python
plt.plot(x1,x2,'o');
```

<br>
<br>
And here is the same bivariate data plotted using seaborn:

```python
sns.jointplot(x="Weight", y="Horsepower", data=cdata, kind = 'reg');
```

<br>
In seaborn, there are up to five different "kinds"
<br>
<br>
Kind of plot to draw:
<br>
kind : { scatter | reg | resid | kde | hex }, optional

you can change the type of graph altogether and use a different style such as a boxplot

```python
plt.figure(figsize=(12, 10))
sns.boxplot(x = "Year", y = "Horsepower", data = cdata, fliersize = "8");
```

```python
plt.figure(figsize=(14, 12))
sns.stripplot(x = "Cylinders", y = "Horsepower", data = cdata, jitter = True, hue= "Origin", 
              dodge = True, size = 8, linewidth = 1);
```

Here is an example of using violin plots for univariate compared to bivariate data:

```python
sns.violinplot(cdata.Weight);
```

```python
sns.violinplot(x = "Origin", y = "Horsepower", data = cdata);
```

```python
plt.figure(figsize = (14,10))
sns.swarmplot(x = "Cylinders", y = "Horsepower", data = cdata, hue="Origin", size=8);
```

use seaborn, we can also make multivariate distribution plots with pairplot

```python
sns.pairplot(cdata, hue="Origin");
```
