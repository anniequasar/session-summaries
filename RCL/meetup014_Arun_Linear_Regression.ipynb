```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
X = np.arange(1,21)
X
```

```python
np.random.rand(20) *2
```

```python
y = 2*X +5  + np.random.rand(20)*2
y
```

```python
def loss(X,y,w):
    y_pred = w*X
    lo = (y-y_pred)**2
    return np.mean(lo)
```

```python
loss(X,y, w=2)
```

```python
w_list = np.arange(-3,3.5,0.5)
w_list
```

```python
loss_list = []
for w in w_list:
    loss_val = loss(X,y,w)
    loss_list.append(loss_val)
```

```python
loss_list
```

```python
plt.scatter(w_list, loss_list)
plt.xlabel("W")
plt.ylabel("Average loss")
```
